13) Mobile_phone: Any mobile phone

14) Work_phone: Any work phone

15) Phone: Any phone number

16) EMAIL_ID: Any email ID

17) Type_Occupation: Occupation

18) Family_Members: Family size

19 )Label: 0 is application approved and 1 is application rejected.

```python
# Add a column 'Unemployed' and initialize with 0
df['Unemployed'] = 0

# Make negative values in 'Employed_days' positive
df['Employed_days'] = df['Employed_days'].abs()

# Move positive values from 'Employed_days' to 'Unemployed'
positive_values = df['Employed_days'][df['Employed_days'] > 0]
df.loc[df['Employed_days'] > 0, 'Unemployed'] = positive_values

# Set all positive values in 'Employed_days' to 0
df.loc[df['Employed_days'] > 0, 'Employed_days'] = 0

df
```

```python
null_income_rows = df[df['Annual_income'].isnull()]

# Print rows with null values in 'Annual_income' column
print("Rows with null values in 'Annual_income' column:")
print(null_income_rows)
```

```python
# Group data by 'Type_Occupation' and calculate the mean of 'Annual_income'
occupation_mean_income = df.groupby('Type_Income')['Annual_income'].mean()

# Print unique values of 'Type_Occupation' and their mean annual incomes
print("Unique values of 'Type_Occupation' and their mean annual incomes:")
for occupation, mean_income in occupation_mean_income.items():
    print(f"{occupation}: {mean_income}")
```

```python
occupation_mean_income_dict = {
    'Commercial associate': 236101,
    'Pensioner': 154012,
    'State servant': 212142,
    'Working': 181824
}

# Fill null values in 'Annual_income' based on 'Type_Occupation'
for occupation, mean_income in occupation_mean_income_dict.items():
    df['Annual_income'].fillna(mean_income, inplace=True)

# Verify the DataFrame after filling null values
df
```

```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
# Initialize the LabelEncoder
label_encoder = LabelEncoder()
# Fit and transform the 'Size' column
df['Car_Owner(encoded)'] = label_encoder.fit_transform(df['Car_Owner'])
df['Propert_Owner(encoded)'] = label_encoder.fit_transform(df['Propert_Owner'])
df['Type_Income(encoded)'] = label_encoder.fit_transform(df['Type_Income'])
df['EDUCATION(encoded)']  = label_encoder.fit_transform(df['EDUCATION'])
df['Housing_type(encoded)']  = label_encoder.fit_transform(df['Housing_type'])
df['Marital_status(encoded)']=label_encoder.fit_transform(df['Marital_status'])
df['Type_Occupation(encoded)']=label_encoder.fit_transform(df['Type_Occupation'])
```

```python
from sklearn.model_selection import train_test_split

# Splitting the dataset into features (X) and target (y)
X = df[['Car_Owner(encoded)', 'Propert_Owner(encoded)', 'CHILDREN', 'Annual_income',
        'Type_Income(encoded)', 'EDUCATION(encoded)', 'Marital_status(encoding)', 'Housing_type(encoded)',
         'Employed_days', 'Mobile_phone', 'Work_Phone',
        'Phone', 'EMAIL_ID', 'Type_Occupation(encoding)', 'Family_Members']]
y = df['label']

# Splitting the dataset into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100)
```

```python
# Create a mask to identify rows with negative values in the training data
negative_mask = (X_train < 0).any(axis=1)

# Filter the training data to include only rows with negative values
rows_with_negative_values = X_train[negative_mask]

# Print the rows containing negative values
print("Rows containing negative values in the training data:")
print(rows_with_negative_values)
```
